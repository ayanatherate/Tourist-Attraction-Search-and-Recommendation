<h2> To run this project, </h2>

1) click on 'Open In Collab' in the project README

2) Run all the code cells one by one

3) Enter your search location as prompted in the 'location' parameter. Enter search radius as desired.
   <br><br>
   

4) Retrieve you desired recommendations based on a Tourist Attraction returned in (3)
   <br><br>
   Enter the name of the spot in the 'provide_recommendations' function, the first parameter would be 
   the name.
